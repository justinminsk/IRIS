```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pygal
#import Librarys
```

```python
iris = pd.read_csv('Iris.csv')
iris.head()
#import data
```

```python
iris.columns
#see column names
```

```python
count_speices = pd.value_counts(iris['Species'], sort = True).sort_index()
count_speices
#count of speices
```

```python
bar = pygal.Bar()
bar.title = 'Iris count'
bar.x_labels = map (str,('Iris-setosa', 'Iris-versicolor', 'Iris-virginica')) #note how the code here is different when we use names
bar.add('Name', count_speices)
bar.render_to_file('bar_chart.svg')
#bar graph comparing the speices
```

```python
color_dict = {'Iris-setosa':'blue', 'Iris-virginica':'red', 'Iris-versicolor':'black'}
#create a color dictionary
iris['Species'].value_counts() 
#count of species in the data
```

```python
iris.plot(kind = "scatter", x = "SepalLengthCm", y = "SepalWidthCm", c = [color_dict[i] for i in iris['Species']])
plt.legend([color_dict])
#plot Sepal length by Sepal width
#blue is setosa. red versicolor, and black virginica
plt.show()
```

```python
Sepal = iris['SepalLengthCm'] * iris['SepalWidthCm']
Petal = iris['PetalLengthCm'] * iris['PetalWidthCm']
#create varibles for the area of the petal and sepal
```

```python
size = (Sepal + Petal)
#create a size varible
s1 = pd.Series(size, name = 'Sepal+Petal')
s2 = pd.Series(Sepal, name = 'Sepal')
s3 = pd.Series(Petal, name = 'Petal')
iris_2 = pd.concat([iris, s1, s2, s3], axis=1)
#add our three new varibles to the data
iris_2.columns
#double check it added
```

```python
iris_2.plot(kind = "scatter", x = "Sepal", y = "Petal", c = [color_dict[i] for i in iris['Species']])
plt.legend([color_dict])
#see Sepal area compared to Petal area
#blue is setosa. red versicolor, and black virginica
plt.show()
```

```python
iris_2.plot(kind = "scatter", x = "Sepal+Petal", y = "Sepal", c = [color_dict[i] for i in iris['Species']])
plt.legend([color_dict])
iris_2.plot(kind = "scatter", x = "Sepal+Petal", y = "Petal", c = [color_dict[i] for i in iris['Species']])
plt.legend([color_dict])
#compare the combined area by only one part of the flower
#blue is setosa. red versicolor, and black virginica
plt.show()
```

```python
print('This shows that sepal makes up more of the plant then the petal.') 
print('We also can see that the iris-setosa has very small petals compared to the other two flowers.')
print("This means that looking at the area of sepal will make less variance compared to looking at petal area, ")
print('or the combined area of the two. We could also see if in the other two flowers if petal') 
print('area and sepal area correlate or if they are completely independent.')
```

```python
iris_2.plot(kind = "scatter", x = "SepalLengthCm", y = "PetalLengthCm", c = [color_dict[i] for i in iris['Species']])
plt.legend([color_dict])
iris_2.plot(kind = "scatter", x = "SepalWidthCm", y = "PetalWidthCm", c = [color_dict[i] for i in iris['Species']])
plt.legend([color_dict])
#Comparing length of sepal and petal in the first graph and width in the second
#blue is setosa. red versicolor, and black virginica
plt.show()
```

```python
print('Thinking about petals vs sepal I decided to see if lengths correlate')
print('more then the area as you can easily see widths do not correlate at all.')
```
